## Task 1: Data Quality Assessment
Assessment of data quality and completeness in preparation for analysis.

The client provided KPMG with 3 datasets:

* Customer Demographic 
* Customer Addresses
* Transactions data in the past 3 months

```python
# import the libraries and the packages
import numpy as np
import scipy.stats as stats
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read in the excel datasets and put into a pandas dataframe
trans, newCust, cust_demo, cust_addr = None, None, None, None
with pd.ExcelFile("/content/KPMG_VI_New_raw_data_update_final.xlsx") as reader:
    trans = pd.read_excel(reader, sheet_name='Transactions')
    new_cust = pd.read_excel(reader, sheet_name='NewCustomerList')
    cust_demo = pd.read_excel(reader, sheet_name='CustomerDemographic')
    cust_addr = pd.read_excel(reader, sheet_name='CustomerAddress')
```

### 1.1 Customer Demographics

```python
cust_demo.head()
```

```python
cust_demo.shape
```

```python
cust_demo.info()
```

```python
# count rows of missing data for each column
missing_number=cust_demo.isnull().sum().sort_values(ascending=False)

# Calculate percentage of null value
missing_percentage=missing_number*100/len(cust_demo)
missing_info=pd.concat([missing_number,missing_percentage],axis=1,keys=['missing number','missing percentage'])
missing_info.head(10)
```

```python
# Drop the columns with many missing values
drop_threshold = cust_demo.shape[0]*0.5
cust_demo = cust_demo.dropna(thresh=drop_threshold, how='all', axis='columns').copy()
```

```python
# Checking for Duplicates in customer_id column
print(cust_demo.duplicated(['customer_id']).sum())
```

```python
# drop the column with corrupted data
del cust_demo['default']
```

```python
# make sure the ID only contain legitimate digits
cust_demo['customer_id'] = cust_demo['customer_id'].astype('int64')
```

```python
# make sure that gender is a category
cust_demo['gender'] = cust_demo['gender'].astype('category')
```

```python
# check the entered data
cust_demo['gender'].value_counts()
```

```python
# replace all the misspelt or differently formatted data in the gender column
cust_demo['gender'].replace({'M' : 'Male', 'F' : 'Female', 'Femal' : 'Female'}, inplace=True)
```

```python
# check the entered data
cust_demo['gender'].value_counts()
```

```python
# check the frequency of different job titles
cust_demo['job_title'].value_counts()
```

```python
# fill the missing values using the backward filling method
cust_demo['job_title']= cust_demo['job_title'].fillna(method='bfill')
```

```python
# check the frequency of job industries
cust_demo['job_industry_category'].value_counts()
```

```python
# fill the missing values using the forward filling method
cust_demo['job_industry_category']= cust_demo['job_industry_category'].fillna(method='ffill')
```

```python
# make DOB column in the date format
cust_demo['DOB'] = cust_demo['DOB'].dt.date
```

```python
# convert values of deceased_indicator column to boolean data type
f = {'N': False, 'Y': True}
cust_demo['deceased_indicator'] = cust_demo['deceased_indicator'].map(f).fillna(cust_demo['deceased_indicator'])
```

```python
# convert values of owns_car column to boolean data type
g = {'No': False, 'Yes': True}
cust_demo['owns_car'] = cust_demo['owns_car'].map(g).fillna(cust_demo['owns_car'])
```

```python
np.nanmedian(cust_demo['tenure'])
```

```python
# filling in the missing value for tenure with median value
cust_demo['tenure'].fillna(value = 11.0, inplace = True)
```

```python
#check the dataset after cleaning
cust_demo.info()
```

```python
cust_demo.head(10)
```

```python
#print to csv file after the cleaning
cust_demo.to_csv('./cust_demo_cleaned.csv',index=False)
```

### 1.3 Transactions

```python
trans.head()
```

```python
trans.info()
```

```python
# count rows of missing data for each column
missing_number=trans.isnull().sum().sort_values(ascending=False)

# Calculate percentage of null value
missing_percentage=missing_number*100/len(trans)
missing_info=pd.concat([missing_number,missing_percentage],axis=1,keys=['missing number','missing percentage'])
missing_info.head(10)
```

```python
#sort the columns by customer id and transaction date
trans = trans.sort_values(['customer_id', 'transaction_date'])
```

```python
trans['online_order'].value_counts()
```

```python
# fill the missing data using the back-fill method
trans['online_order'] = trans['online_order'].fillna(method='bfill')
```

```python
trans['brand'].value_counts()
```

```python
# fill the missing data using the forward-fill method
trans['brand'] = trans['brand'].fillna(method='ffill')
```

```python
trans['product_line'].value_counts()
```

```python
# fill the missing data using the back-fill method
trans['product_line'] = trans['product_line'].fillna(method='bfill')
```

```python
trans['product_class'].value_counts()
```

```python
# fill the missing data using the back-fill method
trans['product_class'] = trans['product_class'].fillna(method='bfill')
```

```python
trans['product_size'].value_counts()
```

```python
# fill the missing data using the forward-fill method
trans['product_size'] = trans['product_size'].fillna(method='ffill')
```

```python
# work out the average cost (to 2 d.p.) for each customer
pd.options.display.float_format = '{:,.2f}'.format
customerStandardCost = trans.groupby(['customer_id']).mean().round(decimals=2)
customerStandardCost
```

```python
# fill the missing data with the average costs for that customer
trans['standard_cost'] = trans['standard_cost'].fillna(customerStandardCost.standard_cost)
```

```python
# fill the missing data with the average first sold date
trans['product_first_sold_date'] = trans['product_first_sold_date'].fillna(trans['product_first_sold_date'].mean())
```

```python
trans.head()
```

```python
trans.info()
```

```python
#print to csv file after the cleaning
trans.to_csv('./trans_cleaned.csv',index=False)
```

```python
# count rows of missing data for each column
missing_number=new_cust.isnull().sum().sort_values(ascending=False)

# Calculate percentage of null value
missing_percentage=missing_number*100/len(new_cust)
missing_info=pd.concat([missing_number,missing_percentage],axis=1,keys=['missing number','missing percentage'])
missing_info.head(10)
```
